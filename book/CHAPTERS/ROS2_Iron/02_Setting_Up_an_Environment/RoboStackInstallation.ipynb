# Robostack

Robostack is a bundling of ROS2 that is available to install for Windows, Mac, and Linux on Conda.

Robostack is easiest to install, because it will install the more annoying ROS2 dependencies like OpenSSL through Conda. It is easy to integrate more complex python libraries for computer vision and machine learning, but that is beyond the scope of these tutorials.

Installation instructions from Robostack can be found [here](https://robostack.github.io/GettingStarted.html), but the instructions are compiled here as well.

Robostack provides both ROS and ROS2 packages through Conda. Setup is the same for installing both.

To set up your environment, run the following commands:

```bash
mamba create -n ros_env python=3.11
mamba activate ros_env

# remove the defaults channel just in case, this might return an error if it is not in the list which is ok
conda config --env --remove channels defaults
# this adds the conda-forge channel to the new created environment configuration 
conda config --env --add channels conda-forge
# and the robostack-staging channel
conda config --env --add channels robostack-staging
# if you want to use Jazzy, also add its own channel:
conda config --env --add channels robostack-jazzy
```

Now we can download ROS from Conda!

```bash
# Install ros-noetic into the environment (ROS1)
mamba install ros-noetic-desktop

# Install ros-humble into the environment (ROS2 Humble)
mamba install ros-humble-desktop

# Install ros-jazzy into the environment (ROS2 Jazzy)
mamba install ros-jazzy-desktop
```

Press `Y` and enter when prompted.

Then, we can launch our environment.

```bash
mamba deactivate
mamba activate ros_env
```

### Test ROS2

We can test ROS2 by launching `rviz2`, a visualization program for ROS2.

If the prompt says `ros_env`, we are good to proceed. Otherwise, activate your environment again with `mamba activate ros_env`.

We can also check that the networking works by running two nodes at once. Open a second miniforge terminal, and activate your environment.


In one terminal, run:
```bash
ros2 run demo_nodes_cpp talker
```

In the other terminal, run:
```bash
ros2 run demo_nodes_cpp listener
```

You should see output from the talker in your listener terminal!
