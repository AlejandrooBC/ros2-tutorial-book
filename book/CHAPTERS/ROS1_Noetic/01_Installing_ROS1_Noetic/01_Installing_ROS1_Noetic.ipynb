# Installing ROS1 Noetic

### Allow Ubuntu Repositories

To Install ROS1 Noetic, Ubuntu requires the user to alow the "restricted," "universe," and "multiverse" repositories. This is due to the dependencies ROS1 Noetic uses. To do this, run the following commands to add the repositories, then update the sources:

```bash
$ sudo add-apt-repository "deb http://archive.ubuntu.com/ubuntu $(lsb_release -sc) main universe restricted multiverse"
$ sudo apt-get update
```

**Note:** if you get an error due to a public key not being available, run the following command, but replace KEY_NAME with the public key that the program is missing:

image.png

```bash
$ sudo apt-key adv --keyserver keyserver.ubuntu.com --recv-keys KEY_NAME
```

### Add ROS Packages

To properly gain access to the ROS repos, your machine needs to use packages.ros.org as a source:

```bash
$ sudo sh -c 'echo "deb http://packages.ros.org/ros/ubuntu $(lsb_release -sc) main" > /etc/apt/sources.list.d/ros-latest.list'
```

### Install Curl

If you do not have curl, make sure to install curl with:
<br>```$ sudo apt install curl```

### Set up keys:

Keys are needed for package authentication. To add the keys for noetic use the command:

```bash
$ curl -s https://raw.githubusercontent.com/ros/rosdistro/master/ros.asc | sudo apt-key add -
```

### Install ROS1 Noetic

Before installing the packages for Noetic, make sure that your Ubuntu packages are updated:

```bash
$ sudo apt update
```

For this tutorial, we will use **ros-noetic-desktop-full**. To do that install the package with:<br>


```bash
$ sudo apt install ros-noetic-desktop-full
```

This tutorial also uses Python. In order to use proper dependencies, also install pythons dependencies with:<br>
```bash
$ sudo apt install python3-rosdep python3-rosinstall python3-rosinstall-generator python3-wstool build-essential
```

### Integrating Bash into your Environment

To use ROS on your shell each time you start up, the ROS script needs to be sourced:<br>
```bash
$ source /opt/ros/noetic/setup.bash
```
The recommendation for this is to add the above line to whatever shell you use.<br>
```bash
$ echo "source /opt/ros/noetic/setup.bash" >> ~/.bashrc
$ source ~/.bashrc
```

### Initialize ROS Dependencies

To use ROS, rosdep is required in order to use core components of ROS. To do this, also install **python3 rosdep**

```bash
$ sudo apt install python3-rosdep
$ sudo rosdep init
$ rosdep update
```

### Ensure Your Packages were Installed Properly

To make sure you installed ROS properly, run: 

```bash
$ printenv | grep ROS
```
